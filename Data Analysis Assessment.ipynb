```python
import matplotlib.pyplot as plt

data.select_dtypes(include=['number']).hist(figsize=(20, 8), bins=30, edgecolor='black')
plt.suptitle("Distribution of Numerical Features", fontsize=14)
plt.show()
```

```python
import seaborn as sns


categorical_cols = data.select_dtypes(include=['object']).columns


plt.figure(figsize=(20, 8))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(3, 4, i)
    sns.countplot(x=data[col], palette="viridis")
    plt.xticks(rotation=45)
    plt.title(f"Count of {col}")
    
plt.tight_layout()
plt.show()
```

# Loan status Based on Previous Campaign 

```python

repeat_customers = data.groupby("poutcome")["loan"].value_counts().unstack()


repeat_customers.plot(kind="bar", stacked=True, colormap="plasma", figsize=(8, 5))
plt.title("Loan Status Based on Previous Campaign Outcome", fontsize=14)
plt.xlabel("Previous Campaign Outcome", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Loan Status", labels=["No", "Yes"])
plt.show()
```

# Job Roles Based On Eduction Level

```python

plt.figure(figsize=(12, 6))
sns.countplot(data=data, x="education", hue="job", palette="viridis")
plt.title("Job Roles Based on Education Level")
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Job Role", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Loan Status By Martial Status

```python


loan_marital_counts = data.groupby(["marital", "loan"]).size().unstack()


plt.figure(figsize=(8, 5))
loan_marital_counts.plot(kind="bar", stacked=True, colormap="viridis", figsize=(8, 5))
plt.title("Loan Status by Marital Status", fontsize=14)
plt.xlabel("Marital Status", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Loan", labels=["No", "Unknown", "Yes"])
plt.show()


loan_marital_counts

```

# conclusion for research
Married people have the highest number of loans (3,773), but also the highest number of people without loans (20,567)

Singles have 1,788 people with loans, while most (9,500) do not have one.

Divorced individuals have 675 with loans, which is relatively lower compared to other groups.

The Unknown category has very few cases.

# Bivariate Analysis

```python
numeric_df = data.select_dtypes(include=['number'])

# Plot the heatmap
plt.figure(figsize=(15, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=data["y"], y=data["age"], palette="Set2")
plt.title("Age Distribution by Target Variable (y)")
plt.show()
```

# conclusion

1.Loan vs. Marital Status

Married people have the highest number of loans, but they also form the largest group in the dataset.
Single and divorced individuals have fewer loans in comparison.

2.Age vs. Balance

No strong correlation was found between age and account balance, indicating that savings do not increase linearly with age.

3.Education vs. Job

People with university degrees are more likely to work in admin or management jobs.
People with primary education are often found in blue-collar jobs.

4.Campaign vs. Previous Outcome

If a client was contacted in a previous campaign and accepted, they are more likely to accept again.
Clients who previously declined tend to decline again.

5.Month vs. Loan Approval

Most loans were approved in May, while fewer approvals happened towards the end of the year.
